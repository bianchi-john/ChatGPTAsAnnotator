```python
import json

def stringa_to_json(stringa):
    # Dividere la stringa in righe e rimuovere eventuali spazi bianchi in eccesso
    righe = [line.strip() for line in stringa.split('\n') if line.strip()]

    # Inizializzare un dizionario per contenere i dati
    dati = {}

    # Variabile per tenere traccia della chiave corrente
    current_key = None

    # Iterare sulle righe estrarre le chiavi e i valori
    for riga in righe:
        # Se la riga inizia con "Q", allora è una nuova chiave
        if riga.startswith("Q"):
            current_key = riga.split(" ")[0].replace(':', ',')
            dati[current_key] = {}
        # Altrimenti, se la riga non inizia con "Q" e c'è una chiave corrente, allora è una risposta
        elif current_key:
            valore = riga.split(".")[0].strip().split()[-1]
            dati[current_key] = valore

    # JSON risultante
    json_string = json.dumps(dati, indent=4)
    return json_string

# Esempio di utilizzo della funzione
stringa = """
Q1.1 What type of news article are you reviewing? (Check one):
    1.  Straight news, or 'hard news'

Q1.2: How accurately does the story’s headline describe the content of the story? (Check one):
    4.  Extremely accurately

Q1.3: Does the story’s headline contain any of the following elements? (Check all that apply):
    1.  A proper noun

Q1.3.1: If the headline contains a quotation, does that same quotation appear in the content of the story? (Check one):
    -1. No quotation present

Q1.4: Does the article begin with a fact-based lead? (Check one):
    2.  Yes

Q1.5 Please rate the degree of bias in the article copy (Check one):
    4.  Entirely unbiased

Q1.6 Does the article use sensationalised language? (Check one):
    3.  Mainly neutral

Q1.7 Does this story negatively target a specific group or individual? (Check one):
    1.  No

Q2.8 How much information is provided in the article’s byline? 
    3.  There is a full name for the author/authors and/or the specific newswire service name

Q2.9.1 Can you determine the date of the event covered by the article? (Check one):
    2.  Yes

Q2.9.2 If the answer to Q2.9.1 is yes (2. Yes), enter the date of the event being reported (DD/MM/YYYY):
    03/07/2021

Q2.10 Is the story covering a news event or development that occurred within 30 days prior to the article’s publication date? (Check one):
    1.  No
"""

json_result = stringa_to_json(stringa)
print(json_result)
```
