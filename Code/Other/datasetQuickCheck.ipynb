```python
import pandas as pd

df = pd.read_csv('../merged_checked.csv')

df
```

```python
# Filtra gli articoli annotati sia da Manuel che da ChatGPT4AsAnnotator
df = df[
    (df['annotator'] == 'Manuel') | 
    (df['annotator'] == 'ChatGPT4AsAnnotator')
]
df


# Conta quante volte appare ciascun articolo
article_counts = df['id'].value_counts()

# Filtra il DataFrame mantenendo solo le righe degli articoli che compaiono esattamente due volte
df = df[df['id'].isin(article_counts[article_counts == 2].index)]

# Ora df_filtered contiene solo le righe degli articoli che compaiono due volte
print(df)
```

```python
print('Manuel')
manuel = df[df["annotator"] == "Manuel"]
print(manuel['Q2.10'].value_counts())

print('---------------------------')
print('---------------------------')

gpt = df[df["annotator"] == "ChatGPT4AsAnnotator"]
print('ChatGPT')
print(gpt['Q2.10'].value_counts())
```
