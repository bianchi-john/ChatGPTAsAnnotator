# The difference between annotations and original annotations is that in Annotations some things have been cleaned by hand 

```python
import sys
sys.path.append('../modules')

import pandas as pd
from sklearn.metrics import cohen_kappa_score,accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from noiseCleaner import noiseCleaner
from deleteNotCoerent import deleteNotCoerent
from cleanAndCast import cleanAndCast
from seeDistinctValues import seeDistinctValues


#Leggo il file in output di chatGPT
chatGPTOutpt = pd.read_csv('../Query/output/newVersion.csv')
#Leggo le annotazioni parsate degli annotatori
annotazioni = pd.read_csv('../Annotations/Annotations.csv')
```

```python
# Rinomino le colonne nel file di annotazioni originali

annotazioni.drop(['url', 'URL(*)', 'Q2.11 Other media', 'Q2.12 News presented type', 'Q2.13 Visual bias', 'Q2.8 Attribution', 'Q2.9.1 Date', 'Q2.9.2 Exact Date', 'Q2.10 event 30 days', 'label'], axis=1, inplace=True)
annotazioni.fillna('', inplace=True)

# Rinomina le colonne utilizzando solo i numeri
annotazioni = annotazioni.rename(columns={
    'Q.1.1 Type of news': 'Q1.1',
    'Q.1.2 Accuracy': 'Q1.2',
    'Q.1.3 Headline': 'Q1.3',
    'Q1.3.1 Quotation': 'Q1.3.1',
    'Q.1.4 Fact-based lead': 'Q1.4',
    'Q.1.5 Bias': 'Q1.5',
    'Q.1.6 Sensational language': 'Q1.6',
    'Q1.7 Negative target': 'Q1.7',
    'Q1.7.1 If the article negatively targets a specific group, please check the types of targeting from the options below': 'Q1.7.1',
    'Q1.7.1.2 Please specify the type': 'Q1.7.1.2'
})

chatGPTOutpt.drop(['label'], axis=1, inplace=True)
```

```python
############### CONTROLLO SE CI SONO VALORI STRANI #################
noiseCleaner(annotazioni)
print('*********************************')
print('*********************************')
print('*********************************')
noiseCleaner(chatGPTOutpt)
```

```python
############ CONTROLLO I SINGOLI VALORI DISTINTI PER COLONNA ################
seeDistinctValues(annotazioni)
print('*********************************')
print('*********************************')
print('*********************************')
seeDistinctValues(chatGPTOutpt)
```

```python
########## CASTO I TIPI E PULISCO ##############

annotazioni = cleanAndCast(annotazioni)
chatGPTOutpt = cleanAndCast(chatGPTOutpt)
```

```python
############ CONTROLLO I SINGOLI VALORI DISTINTI PER COLONNA ################

seeDistinctValues(annotazioni)
print('*********************************')
print('*********************************')
print('*********************************')
seeDistinctValues(chatGPTOutpt)
```

```python
########## Elimino le righe a cui mancano risposte obbligatorie ##########

annotazioni = deleteNotCoerent(annotazioni, 'AnnotatoriUmani')
print('*********************************')
chatGPTOutpt = deleteNotCoerent(chatGPTOutpt, 'ChatGPT')
```
