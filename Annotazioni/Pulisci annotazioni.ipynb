```python
import pandas as pd
import numpy as np

df = pd.read_csv('Annotazioni.csv')
df.drop(['url', 'URL(*)', 'Q1.7.1 If the article negatively targets a specific group, please check the types of targeting from the options below', 'Q2.9.2 Exact Date', 'Q2.11 Other media', 'Q2.12 News presented type', 'Q2.13 Visual bias'], axis=1, inplace=True)

df
```

```python
# Pulizie generali

df = df.applymap(lambda x: x.rstrip(';') if isinstance(x, str) else x)
```

```python
def Q1_1(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '3', '4', '5', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q1_2(df, column):
    df[column] = df[column].astype(str)
    df[column] = df[column].str.replace('.0', '')
    df[column] = df[column].str.replace('nan', '')
    valid_values = ['1', '2', '3', '4', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q1_3(df, column):
    df[column] = df[column].astype(str)
    df[column] = df[column].apply(lambda x: '' if not x.replace(';', '').isdigit() else x)

def Q1_31(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q1_4(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q1_5(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '3', '4', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q1_6(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '3', '4', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q1_7(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q2_8(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '3', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q2_91(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

def Q2_10(df, column):
    df[column] = df[column].astype(str)
    valid_values = ['1', '2', '']
    df[column] = df[column].apply(lambda x: x if x in valid_values else '')

Q1_1(df, 'Q.1.1 Type of news')

Q1_2(df, 'Q.1.2 Accuracy')

Q1_3(df, 'Q.1.3 Headline')

Q1_31(df, 'Q1.3.1 Quotation')

Q1_4(df, 'Q.1.4 Fact-based lead')

Q1_5(df, 'Q.1.5 Bias')

Q1_6(df, 'Q.1.6 Sensational language')

Q1_7(df, 'Q1.7 Negative target')

Q2_8(df, 'Q2.8 Attribution')

Q2_91(df, 'Q2.9.1 Date')

Q2_10(df, 'Q2.10 event 30 days')


```

```python
# CASTO I TIPI

# Sostituisci tutte le stringhe vuote con 0 in tutto il dataframe
df.replace(r'^\s*$', 0, regex=True, inplace=True)

# Columns to convert to integers
columns_to_convert = ['Q.1.1 Type of news', 'Q.1.2 Accuracy', 'Q1.3.1 Quotation', 'Q.1.4 Fact-based lead', 'Q.1.5 Bias', 'Q.1.6 Sensational language', 'Q1.7 Negative target', 'Q2.8 Attribution', 'Q2.9.1 Date', 'Q2.10 event 30 days']

for col in columns_to_convert:
    df[col] = df[col].fillna(0).astype(int)

df['Q.1.3 Headline'] = df['Q.1.3 Headline'].str.split(';')

```
